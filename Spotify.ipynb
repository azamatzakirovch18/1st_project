faqat object va float data type lar bor ekan

```python
floats = []
objects = []
columns = df.columns
```

```python
for i in columns:
    if df[i].dtype == 'float64':
        floats.append(i)
    else:
        objects.append(i)
```

```python
print(f'len of floats {len(floats)}')
print(f'len of objects {len(objects)}')
```

float data type dan boshlayman ! nega ? chunki ko'p

```python
fig, ax = plt.subplots()
for i in floats:
    f = ax.bar(i, df[i].isnull().sum() / 1140)
    ax.bar_label(f,rotation=90,padding=10)
    
ax.set_ylim(0,20)
ax.set_facecolor('#f0f0f0')
ax.grid(True, which = 'major',axis='y',linestyle=':',color='r')
plt.xticks(rotation=90)

ax.set_ylabel('Percentages')
ax.set_title('Percentage of missed values')
ax.set_xlabel('Columns')
plt.show()
```

```python
for i in floats:
    mean = df[i].mean()
    df[i] = df[i].fillna(mean)
```

```python
fig, ax = plt.subplots()
for i in floats:
    f = ax.bar(i, df[i].isnull().sum() / 1140)
    ax.bar_label(f,rotation=90,padding=10)
    
ax.set_ylim(0,20)
ax.set_facecolor('#f0f0f0')
ax.grid(True, which = 'major',axis='y',linestyle=':',color='r')
plt.xticks(rotation=90)
plt.show()
```

floats are done next work is objects

```python
fig, ax = plt.subplots()
for i in objects:
    f = ax.bar(i, df[i].isnull().sum() / 1140)
    ax.bar_label(f,rotation=90,padding=10)
    
ax.set_ylim(0,20)
ax.set_facecolor('#f0f0f0')
ax.grid(True, which = 'major',axis='y',linestyle=':',color='r')
plt.xticks(rotation=90)

ax.set_ylabel('Percentages')
ax.set_title('Percentage of missed values')
ax.set_xlabel('Columns')
plt.show()
```

```python
for i in objects:
    mode = df[i].mode()[0]
    df[i] = df[i].fillna(mode)
```

```python
fig, ax = plt.subplots()
for i in objects:
    f = ax.bar(i, df[i].isnull().sum() / 1140)
    ax.bar_label(f,rotation=90,padding=10)
    
ax.set_ylim(0,20)
ax.set_facecolor('#f0f0f0')
ax.grid(True, which = 'major',axis='y',linestyle=':',color='r')
plt.xticks(rotation=90)

ax.set_ylabel('Percentages')
ax.set_title('Percentage of missed values')
ax.set_xlabel('Columns')
plt.show()
```
